```python
from numpy import asarray #we are using numpy for mathematical function
from numpy import arange
from numpy.random import rand #rand generates random number
from matplotlib import pyplot #for plots 
def objective(x):
    return x**2.0 # x square parabola function
def derivative(x):
    return x*2.0 #derivative
def gradient_descent(objective,derivative,bounds,n_iter,step_size):
    solutions,scores=list(),list()
    solution=bounds[:,0]+rand(len(bounds))*(bounds[:,1]-bounds[:,0])
    for i in range(n_iter):
        gradient=derivative(solution)
        solution=solution-step_size*gradient
        solutions_eval=objective(solution)
        solutions.append(solution)
        scores.append(solutions_eval)
        print('>%d f(%s)=%.5f'%(i,solution,solutions_eval))
    return [solutions,scores]
bounds=asarray([[-1.0,1.0]])
n_iter=30
step_size=0.1
solutions,scores=gradient_descent(objective,derivative,bounds,n_iter,step_size)
inputs=arange(bounds[0,0],bounds[0,1]+0.1,0.1)
results=objective(inputs)
pyplot.plot(inputs,results)
pyplot.plot(solutions,scores,'.-',color='red')
pyplot.show()

```
