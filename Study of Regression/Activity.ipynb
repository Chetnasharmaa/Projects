```python
#pandas are libraries of python which helps you to visualize the data with columns and rows
# Pandas that uses data. We can do a lots of data manipulation like inserting rows, deleting columns, etc
#We can conclude the datas 
import pandas as pd 
import numpy as np #numpy works on arrays. Whenever we wanna create arrays of datas, we use numpy for visualization
import seaborn as sns #seaborn is basically used to create heatmap, visualize datas in form of maps
import matplotlib.pyplot as plt #It helps in visualizing data in forms of graphs (2d, x, y) no z-co-ordinates
```

```python
df=pd.read_csv("titanic.csv", sep="\t")#read the data file , \t for reading values using separator coz
#particular file isnt separated by commas so gotta use \t
```

```python
df.head(10) # head is a function that shows column, rows of the data
```

```python
df.shape # shape says how many columns and rows are there (156 rows, 12 columns)
```

```python
df.isnull().sum() #Age--->30so 30 null values in the age column, etc 
```

```python
sns.heatmap(df.isnull(),cmap="spring")#heatmap-->a form of maps which works over color formation( dark, light) 
#more intensity--> darkness, less-->lighter, cmap==string is a type of heatmap
```

```python
#drop/delete a column
df.drop("Embarked",axis=1,inplace=True) #axis=1 is basically 1 column, inplace=True means in the place 
#you are dropping column, you say True, used to avoid null values, understanding--> no null value, avoid the null values
#If null values are there, then it will affect a larger data, other rows, their values, 
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#drop all null values, null value means nothing, empty ( 0 cant be null value, its a value)
df.dropna(inplace=True)
```
