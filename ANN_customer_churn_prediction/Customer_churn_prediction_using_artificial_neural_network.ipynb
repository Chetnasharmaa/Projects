```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

"""### **2. Import Dataset**"""

df = pd.read_csv('Churn_Modelling.csv')

df.head()

df.info()

df.describe()

"""**Conclusion -**

1. Dataset Shape - (10000, 14)#samples, columns(features)

2. Numerical features - 11 #neural network works with numbers

3. Categorical features - 3

4. Null Values - 0

### **3. Data Preprocessing**

#### **3.1. Label Encoding**

"""

from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()#categorical data into numerical value

df['Geography'] = lb.fit_transform(df['Geography']) #

df['Gender'] = lb.fit_transform(df['Gender']) #

df

df.info()

"""#### **3.2. Feature Selection**"""

df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)

df.shape

"""#### **3.3. Train Test Split**"""

y = df.pop('Exited')#removes the particular column 

X = df

X.shape

y.shape

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)#randomstate u can use 42 as well

"""#### **3.4. Feature Scaling**"""

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)

X_test = sc.fit_transform(X_test)

X_train 
import keras

from keras.models import Sequential

from keras.layers import Dense

from keras.layers import LeakyReLU, PReLU, ELU #these are activation function, we are not using activation function, these are different under ReLu

from keras.layers import Dropout #prevent overfitting=regularization

#initialize the ANN

classifier = Sequential()

# Add the input layer and the first hidden layer

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu',input_dim = 10))#he_uniform is a function to initialize weights, make the weights( this is first input layer so we need to make weight)

# Add the second hidden layer

classifier.add(Dense(units = 6, kernel_initializer = 'he_uniform',activation='relu'))

# Add the output layer

classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))# with sigmoid, use glorot_uniform

# Compile the ANN

classifier.compile(optimizer = 'Adamax', loss = 'binary_crossentropy', metrics = ['accuracy'])#adamax adjusts weight while training, loss=in binary classification, we use this

# Fitting the ANN to the Training set

model_history=classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)#stores training history, loss, epochs

# list all data in history

classifier.summary()#displays network architecture, shows parameters like weights & bias

#total params are total number of weights and biases

"""### **5. Model Evaluation**"""

# Predict the Test set results

Y_pred = classifier.predict(X_test)#prediction of probability for the test set , has an array of probabilities

Y_pred

Y_pred = (Y_pred > 0.5)#true, leave, 0.5 is threshold

Y_pred

# Confusion Matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, Y_pred)

print(cm)

# Calculate the Accuracy

from sklearn.metrics import accuracy_score

score=accuracy_score(Y_pred,y_test)

print(score)
```
