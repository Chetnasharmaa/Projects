```python
#ipynb is for visualization of data (operation and manipulation)
import matplotlib.pyplot as plt 
import numpy as np
from sklearn import datasets, linear_model #sklearn has datasets
from sklearn.metrics import mean_squared_error, r2_score # mean_squared= measure square difference of actual(initial) and predicted value, r2=coefficient(maths, measures how well the model eplains the varaince/randomness(the more creative, the randomness increases, the less random model will be, less good output)
```

```python
diabetes_X,diabetes_y=datasets.load_diabetes(return_X_y=True) #X is capital? To match dataset we write X,y.. Diabetes_X contains features, attributes related to an individual and shows progressions over the year 
```

```python
diabetes_X=diabetes_X[:,np.newaxis,2]#: is basically start of list and going to the 2, its the end, end of the dataset, here bmi 
```

```python
diabetes_X_train=diabetes_X[:-20]#all the rows except the last 20 rows
diabetes_X_test=diabetes_X[-20:]#all the last 20 rows which we didn't take in training sets
diabetes_y_train=diabetes_y[:-20]
diabetes_y_test=diabetes_y[:-20]
```

```python
model=linear_model.LinearRegression() #() is a function
```

```python
model.fit(diabetes_X_train,diabetes_y_train)
```

```python
diabetes_y_pred=model.predict(diabetes_X_test)#over the years
```

```python
print('mean_squared_error:%.2f')#2f stores float value, mean_squared_error has float values, float increases accuracy 2f=423.0 2f=423.00
print('Coefficient:%.2f')
```
