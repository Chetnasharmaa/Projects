```python
from numpy import where
from collections import Counter #collections is used to count handling objects
from sklearn.datasets import make_blobs #generates isotropic Gaussian Blobs for clustering
from matplotlib import pyplot #plot


```

```python
#Defining Dataset
X, y = make_blobs(n_samples=1000, centers=2, random_state=1)
#Dataset shape
print (X.shape, y.shape)
#Summarize observation by Class Label
counter=Counter(y)
print(counter)
#Summarize first few examples
for i in range(10):
  print(X[i],y[i])

```

```python
#Design a plot
for i,j in counter.items():
  row=where(y==i)[0]
  pyplot.scatter(X[row,0],X[row,1],label=str(i))
pyplot.legend()
pyplot.show()
```
