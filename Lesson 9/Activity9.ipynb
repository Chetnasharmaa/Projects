```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score, mean_squared_error

# Load the dataset and handle the empty first column

df = pd.read_csv('xydataset.csv')

print("Original data shape:", df.shape)

print("Original columns:", df.columns.tolist())#converts column to list 

print(df.head()) # print all rows & columns

# Remove the empty first column (index 0) and use columns 1 and 2

df_clean = df.iloc[:, 1:3] # Take columns 1 and 2 (skip the empty first column)
#0 is the first column
print("\nCleaned data:")

print(df_clean.head())

# Extract X and Y values from the cleaned dataframe

X = df_clean.iloc[:, 0].values.reshape(-1, 1) # First actual data column
#reshape x column to a particular array (list), values=picking one one value 
#reshape column is independent variable

y = df_clean.iloc[:, 1].values # Second actual data column

print(f'\nX shape: {X.shape}') # shape gives dimension(how many rows)

print(f'Y shape: {y.shape}')

# Create and fit the model

model = LinearRegression()

model.fit(X, y)

# Make predictions

y_pred = model.predict(X)

# Calculate metrics

R2 = r2_score(y, y_pred)

mse = mean_squared_error(y, y_pred)

rmse = np.sqrt(mse)

# Get coefficients

slope = model.coef_[0] #m, we always do 0 to pick the most initial value 

intercept = model.intercept_ #c

# Print results

print(f'\nRegression Results:')

print(f'Coefficient: {slope:.4f}')

print(f'Intercept: {intercept:.4f}')

print(f'R-square: {R2:.4f}')

print(f'RMSE: {rmse:.4f}')

print(f'Regression equation: y = {slope:.4f}x + {intercept:.4f}')

# Create the plot

plt.figure(figsize=(20, 20)) # screen size, width, height

plt.scatter(X, y, color='blue', alpha=0.6, label='Data')#alpha --> to print scatter in proper/brightness of scatter plots (say) they are just original data sets not predicted

plt.plot(X, y_pred, color='red', linewidth=6, label=f'Regression Line (R² = {R2:.4f})')# for predicted line

plt.xlabel('X Variable')

plt.ylabel('Y Variable')

plt.title('Linear Regression Analysis')

plt.legend()#there are lot of things inside. we use to basically manage all 3/4 features it has. it also mentions about the locations of text. it also handles location of the title as well. frame location, font size, font locations

plt.grid(True, alpha=1)#grid is for grid format

# Add regression equation on the plot

equation_text = f'y = {slope:.4f}x + {intercept:.4f}'

plt.text(0.05, 0.95, equation_text, transform=plt.gca().transAxes,

bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.9),

) # 0.05,0.95 where text should come, transAxes manages the axis(x&y axis), bbox

plt.tight_layout() # This should come before plt.show(), layout of the graph should not leak. what does it mean? in graph lots of datas, and scattering is all over the graph and scatter sometimes get out of the graph

plt.show() # Fixed: added parentheses # very very importance 




```
